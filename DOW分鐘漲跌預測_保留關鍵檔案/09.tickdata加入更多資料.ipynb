```python
import numpy as np 
import pandas as pd 
```

```python
tickdata_df = pd.read_csv("datas/1min_MYM加入5~30分鐘標籤_萬分位漲跌幅.csv")
# tickdata_df = pd.read_csv("datas/tickdata加入答案標籤_2_加入15~20分鐘標籤_萬分位漲跌幅.csv")
tickdata_df = tickdata_df.drop(["Unnamed: 0"],axis=1)
tickdata_df.set_index(["datetime"],inplace=True)
tickdata_df.columns
```

```python
tickdata_df.head()
```

```python
# 將[ "5分鐘後的漲跌幅","10分鐘後的漲跌幅","15分鐘後的漲跌幅","20分鐘後的漲跌幅","5分鐘後的震盪程度","10分鐘後的震盪程度"] 這幾個column挪到最前面
# 假設 data 是你的 DataFrame
# 設定要移動到最前面的 column
columns_to_move = [
    "5分鐘後的漲跌幅",
    "10分鐘後的漲跌幅",
    "15分鐘後的漲跌幅",
    "20分鐘後的漲跌幅",
    "30分鐘後的漲跌幅",
]

# 獲取 DataFrame 中其餘的 column
remaining_columns = [col for col in tickdata_df.columns if col not in columns_to_move]

# 重新排列 column，將指定的 column 移到最前面
tickdata_df = tickdata_df[columns_to_move + remaining_columns]

# 確認新的 column 排列順序
print(tickdata_df.columns)
```

```python
df_技術= tickdata_df.copy()
# 1. VWAP
df_技術["vwap"] = (df_技術["close"] * df_技術["volume"]).cumsum() / df_技術["volume"].cumsum()

# 2. Bollinger Bands (20-period SMA and standard deviation)
window = 20
df_技術["sma_20"] = df_技術["close"].rolling(window).mean()
df_技術["std_20"] = df_技術["close"].rolling(window).std()
df_技術["bollinger_upper"] = df_技術["sma_20"] + (df_技術["std_20"] * 2)
df_技術["bollinger_lower"] = df_技術["sma_20"] - (df_技術["std_20"] * 2)
df_技術["bollinger_bandwidth"] = (df_技術["bollinger_upper"] - df_技術["bollinger_lower"]) / df_技術[
    "sma_20"
]

# 3. RSI (14-period)
delta = df_技術["close"].diff()
gain = (delta.where(delta > 0, 0)).rolling(window=14).mean()
loss = (-delta.where(delta < 0, 0)).rolling(window=14).mean()
rs = gain / loss
df_技術["rsi"] = 100 - (100 / (1 + rs))

# 4. MACD (12, 26, 9 EMA)
df_技術["ema_12"] = df_技術["close"].ewm(span=12, adjust=False).mean()
df_技術["ema_26"] = df_技術["close"].ewm(span=26, adjust=False).mean()
df_技術["macd_line"] = df_技術["ema_12"] - df_技術["ema_26"]
df_技術["signal_line"] = df_技術["macd_line"].ewm(span=9, adjust=False).mean()

# 5. KD-Stochastic (14-period)
low_min_14 = df_技術["low"].rolling(window=14).min()
high_max_14 = df_技術["high"].rolling(window=14).max()
df_技術["%K"] = 100 * ((df_技術["close"] - low_min_14) / (high_max_14 - low_min_14))
df_技術["%D"] = df_技術["%K"].rolling(window=3).mean()

# 6. ATR (14-period)
df_技術["tr"] = np.maximum(
    df_技術["high"] - df_技術["low"],
    np.maximum(
        abs(df_技術["high"] - df_技術["close"].shift(1)), abs(df_技術["low"] - df_技術["close"].shift(1))
    ),
)
df_技術["atr"] = df_技術["tr"].rolling(window=14).mean()

# 丟掉中間計算的columns只保留最終的指標 (可選)
df_技術 = df_技術[
    [
        "vwap",
        "bollinger_upper",
        "bollinger_lower",
        "bollinger_bandwidth",
        "rsi",
        "macd_line",
        "signal_line",
        "%K",
        "%D",
        "atr",
    ]
]

print(df_技術.tail())  # 顯示結果
```

```python
# 合併 df_技術 和 tickdata_df
tickdata_df_加入技術 = tickdata_df.merge(df_技術, on="datetime", how="left")
tickdata_df_加入技術.columns
```
