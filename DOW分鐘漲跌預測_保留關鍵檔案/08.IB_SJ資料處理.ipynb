```python
# 假設 SJ_df_resampled 已經存在，且 ts 欄位是 datetime 格式
def add_minutes_from_open(df):
    # 將 ts 確保是 datetime 格式
    df["ts"] = pd.to_datetime(df.index)

    # 創建一個函數來計算距離開盤時間的分鐘數
    def minutes_from_open(timestamp):
        # 如果時間在 00:00 到 04:00 之間，表示是隔天的交易，要計算從前一天 15:00 開始
        if timestamp.hour < 4:
            # 調整到前一天的 15:00
            open_time = (timestamp - pd.Timedelta(days=1)).replace(
                hour=15, minute=0, second=0
            )
        else:
            # 當天的 15:00
            open_time = timestamp.replace(hour=15, minute=0, second=0)

        # 計算時間差（分鐘）
        minutes = int((timestamp - open_time).total_seconds() / 60)
        return minutes

    # 添加新欄位
    df["距離開盤時間分鐘數"] = df["ts"].apply(minutes_from_open)

    return df


# 處理資料
SJ_IB_merged_df = add_minutes_from_open(SJ_IB_merged_df)
print(SJ_IB_merged_df)
```

```python
def calculate_metrics(df):
    # 確保數據按時間排序
    df = df.sort_index()

    # 計算高低點幅度
    df["高低點幅"] = (df["high"] - df["low"]) / df["open"]

    # 計算收漲跌幅
    df["收漲跌幅"] = (df["close"] - df["open"]) / df["open"]

    # 定義時間週期
    periods = {
        "10分鐘": 2,  # 2個5分鐘週期
        "30分鐘": 6,  # 6個5分鐘週期
        "60分鐘": 12,  # 12個5分鐘週期
        "120分鐘": 24,  # 24個5分鐘週期
    }

    # 計算不同時間週期的收漲跌幅
    for period_name, period in periods.items():
        # 原有的收漲跌幅計算
        previous_close = df["close"].shift(period)
        df[f"近{period_name}的收漲跌幅"] = (
            df["close"] - previous_close
        ) / previous_close

        # 計算高低點幅
        rolling_high = df["high"].rolling(window=period, min_periods=1).max()
        rolling_low = df["low"].rolling(window=period, min_periods=1).min()
        df[f"近{period_name}高低點幅"] = (rolling_high - rolling_low) / df[
            "close"
        ].shift(period)

        # 計算高點到現在的漲跌幅
        df[f"近{period_name}高點到現在的漲跌幅"] = (rolling_high - df["close"]) / df[
            "close"
        ]

        # 計算低點到現在的漲跌幅
        df[f"近{period_name}低點到現在的漲跌幅"] = (rolling_low - df["close"]) / df[
            "close"
        ]

    # 計算台灣-美國高低掛
    df["台灣-美國高低掛"] = df["UDF_close"] - df["close"]

    return df


# 使用範例：
merged_df_加入收漲跌幅和前高低 = calculate_metrics(SJ_IB_merged_df)
merged_df_加入收漲跌幅和前高低
```

```python
import pandas as pd
import numpy as np
from datetime import datetime, time


def calculate_metrics(df):
    # 確保數據按時間排序
    df = df.sort_index()

    # 計算高低點幅度
    df["高低點幅"] = (df["high"] - df["low"]) / df["open"]

    # 計算收漲跌幅
    df["收漲跌幅"] = (df["close"] - df["open"]) / df["open"]

    # 定義時間週期
    periods = {
        "10分鐘": 10,  # 2個5分鐘週期
        "30分鐘": 30,  # 6個5分鐘週期
        "60分鐘": 60,  # 12個5分鐘週期
        "120分鐘": 120,  # 24個5分鐘週期
    }

    # 計算不同時間週期的指標
    for period_name, period in periods.items():
        # 收漲跌幅計算
        previous_close = df["close"].shift(period)
        df[f"近{period_name}的收漲跌幅"] = (
            df["close"] - previous_close
        )*100 / previous_close

        # 計算高低點幅
        rolling_high = df["high"].rolling(window=period, min_periods=1).max()
        rolling_low = df["low"].rolling(window=period, min_periods=1).min()
        df[f"近{period_name}高低點幅"] = (rolling_high - rolling_low)*100 / df[
            "close"
        ].shift(period)

        # 計算高點到現在的漲跌幅
        df[f"近{period_name}高點到現在的漲跌幅"] = (rolling_high - df["close"])*100 / df[
            "close"
        ]

        # 計算低點到現在的漲跌幅
        df[f"近{period_name}低點到現在的漲跌幅"] = (rolling_low - df["close"])*100 / df[
            "close"
        ]

        # 計算前N分鐘成交量總和
        # 使用shift(1)確保不包含當前時間點的成交量
        df[f"前{period_name}成交量總和"] = (
            df["volume"].shift(1).rolling(window=period, min_periods=1).sum()
        )
        df[f"前{period_name}成交量平均"] = (
            df["volume"].shift(1).rolling(window=period, min_periods=1).sum() / period
        )

    # 計算台灣-美國高低掛
    # df["台灣-美國高低掛"] = df["UDF_close"] - df["close"]

    # 計算星期幾
    def get_trading_day(timestamp):
        # 將時間轉換為datetime
        dt = pd.to_datetime(timestamp)
        current_time = dt.time()

        # 如果時間在00:00-04:00之間，使用前一個交易日的星期
        if time(0, 0) <= current_time <= time(4, 0):
            # 往前找到最近的15:00的時間點的星期
            previous_session = dt - pd.Timedelta(hours=dt.hour + 1)  # 確保回到前一天
            previous_session = previous_session.replace(hour=15, minute=0, second=0)
            return previous_session.weekday()
        else:
            # 對於15:00-23:59的時間，使用當天的星期
            return dt.weekday()

    # 添加星期幾欄位 (0=星期一，6=星期日)
    df["星期幾"] = df.index.map(get_trading_day)

    return df


# 使用範例：
merged_df_加入收漲跌幅和前高低含成交量總和和星期 = calculate_metrics(SJ_IB_merged_df)
merged_df_加入收漲跌幅和前高低含成交量總和和星期
```
