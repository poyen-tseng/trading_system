```python
import numpy as np
a = df_headline["llama_score"].copy()
b = df_text["llama_score"].copy()
# 將列表轉為 NumPy 陣列，並將 None 轉為 np.nan
a = np.array(a, dtype=np.float64)
b = np.array(b, dtype=np.float64)

# 找出 a 和 b 中 nan 的位置
nan_mask = np.isnan(a) | np.isnan(b)

# 刪除 nan 的位置
a_cleaned = a[~nan_mask]
b_cleaned = b[~nan_mask]

# 結果
print("新的 a:", a_cleaned.tolist())
print("新的 b:", b_cleaned.tolist())
```

```python
import numpy as np

# 創建包含NaN的陣列
a = np.array(b_cleaned)

# 計算NaN的數量
nan_count = np.sum(np.isnan(a))

print("NaN的數量:", nan_count)
```

```python
import matplotlib.pyplot as plt
x1 = a_cleaned
x2 = b_cleaned

# 依據 a 進行排序
sorted_a, sorted_b = zip(*sorted(zip(x1, x2)))

# 將排序結果轉回 list
x1 = list(sorted_a)
x2 = list(sorted_b)

# 輸出結果
print("新的 a:", x1)
print("新的 b:", x2)

# 繪製圖形
plt.figure(figsize=(12,6))
plt.plot(sorted_a, label="標題", color="blue", marker=".", linestyle="-")
plt.plot(sorted_b, label="內文", color="red", marker=".", linestyle="-")

# 添加標題和軸標籤
plt.title("比較標題和內文")
plt.xlabel("新聞編號")
plt.ylabel("新聞評分")

# 顯示圖例
plt.legend()

# 顯示網格線
plt.grid(True, linestyle="--", alpha=0.6)

# 顯示圖形
plt.show()
```

```python
import numpy as np


def compare_arrays(a, b):
    a = np.array(a)
    b = np.array(b)

    # 絕對值大於 0.1 的索引
    mask_a = np.abs(a) > 0.1
    mask_b = np.abs(b) > 0.1

    # 同時大於 0.1 的索引
    valid_indices = mask_a & mask_b

    # 正負號相同的數量
    same_sign = np.sum((a[valid_indices] * b[valid_indices]) > 0)

    # 正負號相異的數量
    different_sign = np.sum((a[valid_indices] * b[valid_indices]) < 0)

    # 絕對值小於等於 0.1 的索引
    neutral_indices = ~valid_indices
    neutral_same = np.sum(
        (np.abs(a[neutral_indices]) <= 0.1) & (np.abs(b[neutral_indices]) <= 0.1)
    )

    return same_sign, different_sign, neutral_same


# 範例數據
a = x1
b = x2

# 計算結果
same_sign, different_sign, neutral_same = compare_arrays(a, b)
print(f"正負號相同的數量: {same_sign}")
print(f"正負號相異的數量: {different_sign}")
print(f"中性數值相同的數量: {neutral_same}")
print(f"一正一中或一負一中的數量:{len(x2)-different_sign-same_sign-neutral_same}")
print(f"數據總量:{len(x1)},{len(x2)}")
print(f"同質性:{(same_sign+neutral_same)/ len(x2)}")
```
